```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import time
import pandas as pd

resumen= open("paradoja_amistad_100.csv", "a")
resumen.write("{} {} \n".format("Grado_Promedio","Grado_promedio_de_vecinos"))
resumen.close()

for s in range (40):
    print(s)
    print(time.strftime("%H:%M:%S"))
    print("inicio")
    c = np.loadtxt("enlaces" +str(s) +".txt")# lee el archivo que entre columans tiene espacios
    a = len(c) #NÚMERO DE ENLACES
    G = nx.Graph()# creamos el grafo
#############creamos los 10,000 nodos y los enlazamos
    i = 0
    b = 10000
    if i <= b-1:
        G.add_node(i)
        i+=1
    i = 0
    for i in range (a):
        if i == a:
            break
        ta = c[i][0] 
        ya = c[i][1]
        G.add_edge(ta,ya)
##################################
    print(nx.is_connected(G))
    print(nx.number_connected_components(G))
####################################### sub programa para encontrar el nodo de grado máximo
    d = 0
    máximo = 0
    for d in range (10000):
        máximo_ideal = G.degree(d)
        if máximo_ideal > máximo:
            máximo = máximo_ideal
            nodo = d
             
    print(nodo,máximo)###imprime el nodo que tiene grado máximo en la red
####################################### sub programa para aislar el comoponente gigante
    if nx.is_connected(G) == False:
        Grafo2 = nx.node_connected_component(G,nodo)
        tamaño = len(list(Grafo2))
        b = list(Grafo2)
        print(tamaño)
        porcentaje = "{:.2f}".format(tamaño/100)
        print (porcentaje)
    else:
        tamaño = len(list(nx.node_connected_component(G,nodo)))
        porcentaje = "{:.2f}".format(tamaño/100)
        print (porcentaje)
        print("no fue necesario")
    Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
    G0 = G.subgraph(Gcc[0])
    print(nx.number_connected_components(G0))
    G = G0 #####renombramos la red
    print(nx.is_connected(G))
########################################## sub programa para obtener dataframe de paradoja de la amistad
    #Pone en un DataFrame el grado de  cada nodo
    df = pd.DataFrame(list(nx.degree(G)),columns=['Nodo','Grado'])
    #Un DataFrame para el grado promedio de los vecinos 
    lista_2 = []
    for i in list(G):
        neighbors = list(nx.neighbors(G,i))
        nodo = i
        lista_1 = []   
        for j in neighbors:
            #print(nx.degree(G,i)) nos da el grado de cada vecino
            lista_1.append(nx.degree(G,j))
            promedio = sum(lista_1)/len(lista_1) #promedio
        lista_2.append([nodo,promedio])
    df_2 = pd.DataFrame(lista_2,columns=['Nodo','Grado_Promedio_de_Vecinos'])
              
    df_3 = df.merge(df_2, on='Nodo') #unimos las tablas

    a = df_3['Grado'].mean()
    b = df_3['Grado_Promedio_de_Vecinos'].mean()              
####################
    resumen = open("paradoja_amistad_100.csv", "a")
    resumen.write("{} {} \n".format(a,b))
    resumen.close()
    print(time.strftime("%H:%M:%S"))
    print("fin")
    s = s+1
```
