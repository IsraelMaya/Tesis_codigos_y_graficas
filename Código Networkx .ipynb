```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import time

resumen = open("resumen_red.csv", "a")
resumen.write("{} {} {} {} {} \n".format("Kmax","Nodo","Componente_gigante(%)","BC","lij"))
resumen.close()

for s in range (40):
    print(s)
    print(time.strftime("%H:%M:%S"))
    print("inicio")
    c = np.loadtxt("enlaces" +str(s) +".txt")# lee el archivo que entre columans tiene espacios
    a =len(c) #NÚMERO DE ENLACES
    G = nx.Graph()# creamos el grafo
#############creamos los 10,000 nodos y los enlazamos
    i = 0
    b = 10000
    if i <= b-1:
        G.add_node(i)
        i+=1
    i = 0
    for i in range (a):
        if i == a:
            break
        ta = c[i][0] 
        ya = c[i][1]
        G.add_edge(ta,ya)
##################################
    print(nx.is_connected(G))
    print(nx.number_connected_components(G))
####################################### sub programa para encontrar el nodo de grado máximo
    d = 0
    máximo = 0
    for d in range (10000):
        máximo_ideal = G.degree(d)
        if máximo_ideal > máximo:
            máximo = máximo_ideal
            nodo = d
             
    print(nodo,máximo)###imprime el nodo que tiene grado máximo en la red
####################################### sub programa para aislar el comoponente gigante
    if nx.is_connected(G) == False:
        Grafo2 = nx.node_connected_component(G,nodo)
        tamaño = len(list(Grafo2))
        b = list(Grafo2)
        print(tamaño)
        porcentaje = "{:.2f}".format(tamaño/100)
        print (porcentaje)
    else:
        tamaño = len(list(nx.node_connected_component(G,nodo)))
        porcentaje = "{:.2f}".format(tamaño/100)
        print (porcentaje)
        print("no fue necesario")
    Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
    G0 = G.subgraph(Gcc[0])
    print(nx.number_connected_components(G0))
    G = G0 #####renombramos la red
    print(nx.is_connected(G))
    ##################################### sub programa para escribir archivo del componente gigante
    f = open ("enlaces_red" +str(s) +".csv",'a')
    f.write("{} {} \n".format("P(k)","k"))
    f.close()
    lista_nodos=list(G)
    número_de_nodos=len(lista_nodos)
    i=0
    for i in range (número_de_nodos):
        h=lista_nodos[i]
        lista=list(G.adj[h])
        largo=len(lista)
        j=0
        for j in range(largo):
            b=lista[j]
            f=open("enlaces_red" +str(s) +".csv",'a')
            f.write("{} {} \n".format(i,b))
            f.close()
    print("listo archivo 1")
##################################### sub programa para escribir archivo para gephi del componente gigante
    lista_nodos=list(G)
    número_de_nodos=len(lista_nodos)
    i=0
    for i in range (número_de_nodos):
        h=lista_nodos[i]
        lista=list(G.adj[h])
        largo=len(lista)
        j=0
        for j in range(largo):
            b= int(lista[j])
            f=open("enlaces_red" +str(s) +".txt",'a')
            f.write("{} {} \n".format(i,b))
            f.close()
    print("listo archivo 2")
##################################### metricas a escribir
    Cprom=nx.average_clustering(G)
    Cprom="{:.7f}".format(Cprom)
    lij=nx.average_shortest_path_length(G)
    lij="{:.3f}".format(lij)
    BC=nx.betweenness_centrality(G)
###################### guardar en un archivo el nodo y su BC
    h=open("Centralidad" +str(s) +".cvs",'a')
    h.write("{} {} \n".format("Nodo","BC"))
    h.close()
    for key, value in sorted(BC.items(), key=lambda item: item[1]):
        h=open("Centralidad" +str(s) +".cvs",'a')
        h.write("{} {} \n".format(key,value))
        h.close()
####################
    resumen = open("resumen_red.csv", "a")
    resumen.write("{} {} {} {} {} \n".format(máximo,nodo,porcentaje,Cprom,lij))
    resumen.close()
    print(time.strftime("%H:%M:%S"))
    print("fin")
    s=s+1
```
